```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
```

```python
yt_no_jump_02 = np.load('yt_02.npy')
yt_no_jump_025 = np.load('yt_025.npy')
yt_no_jump_03 = np.load('yt_03.npy')

no_jump_prob_02 = np.load('prob_02.npy')
no_jump_prob_025 = np.load('prob_025.npy')
no_jump_prob_03 = np.load('prob_03.npy')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
ax1.plot(yt_no_jump_02, linewidth=2., label=r'$r_0=0.2$')
ax1.plot(yt_no_jump_025, linewidth=2., label=r'$r_0=0.25$')
ax1.plot(yt_no_jump_03, linewidth=2., label=r'$r_0=0.3$')
ax1.plot(np.ones_like(yt_no_jump_02)*1.5, linestyle='dashed', color='black', linewidth=2)
ax1.set_xlabel('Years', fontsize=13)
ax1.set_title(r'Temperature anomaly')
ax1.set_ylim(1, 2)
ax1.legend(frameon=False)

ax2.plot(no_jump_prob_02, linewidth=2., label=r'$r_0=0.2$')
ax2.plot(no_jump_prob_025, linewidth=2., label=r'$r_0=0.25$')
ax2.plot(no_jump_prob_03, linewidth=2., label=r'$r_0=0.3$')
ax2.set_xlabel('Years', fontsize=13)
ax2.set_title(r'No jump probability')
ax2.set_ylim(0, 1.05)
ax2.legend(frameon=False)

fig.suptitle(r'Figure 2: $r_0=0.3$', fontsize=13)
fig.savefig('fig.pdf')
```
