# BHmodified

$z_t$ process:
$$
    dZ_t^1 = \mu_1(z_t^1)dt + \sigma_1(z_t^1)dW^1_t\\
    dZ_t^2 = - \rho(Z_t^2 - \mu_2)dt + \sqrt{Z_t^2}\sigma_2(Z_t^2)dW_t
$$  

```python
# z stationary
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 500
num = 25
W = np.random.normal(0,1, (num, T))
```

```python
ρ = .5
μ_2 = 1.
```

```python
def zGenerator(μ_2, ρ, σ_2, T, num, W):
    Zs = np.zeros((num,T))
    W = np.random.randn(num, T)
    Z = 1
    for year in range(T):
        dZ = -ρ*(Z - μ_2) + np.sqrt(Z)*σ_2*W[:,year]
        Z = dZ + Z
        Zs[:,year]=Z
    return Zs
```

```python
σ_z = .004
σ_2 = np.sqrt(2*σ_z**2*ρ/μ_2)
Zs = zGenerator(μ_2, ρ, σ_2, T, num, W)
```

```python
np.std(Zs), np.mean(Zs)
```

```python
plt.plot(Zs[0])
```

```python
σ_z = .17
σ_2 = np.sqrt(2*σ_z**2*ρ/μ_2)
Zs_goal = zGenerator(1., .5, σ_2, T, num, W)
```

```python
plt.hist(Zs_goal, bins=200, density=1)
plt.show()
```
